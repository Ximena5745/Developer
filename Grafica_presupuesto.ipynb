```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Datos
anos = ["2022", "2023", "2024", "En curso"]
opex = [0, 458118549, 1071756781, 70844000]
capex = [0, 0, 2486524255, 6337013588]
total = [0, 458118549, 3558281036, 6407857588]

# Formatear números con puntos como separadores de miles
def format_number(num):
    return f"{num:,}".replace(",", ".")

# Crear la figura y ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Crear barras apiladas
bar_width = 0.5
x = np.arange(len(anos))
ax.bar(x, opex, bar_width, label='OPEX', color='royalblue')
ax.bar(x, capex, bar_width, bottom=opex, label='CAPEX', color='orange')

# Añadir etiquetas de los valores totales encima de las barras
for i, val in enumerate(total):
    ax.text(x[i], val + (val * 0.02), format_number(val), ha='center', fontsize=10, fontweight='bold')

# Configurar etiquetas y título
ax.set_xticks(x)
ax.set_xticklabels(anos)
ax.set_ylabel("Presupuesto Ejecutado")
ax.set_title("Presupuesto Ejecutado por Año (OPEX y CAPEX)")
ax.legend()

# Crear la tabla con datos
data = list(zip(anos, map(format_number, opex), map(format_number, capex), map(format_number, total)))
df = pd.DataFrame(data, columns=["Año", "OPEX", "CAPEX", "Total"])

# Agregar tabla debajo del gráfico
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='bottom', bbox=[0, -0.5, 1, 0.3])
table.auto_set_font_size(False)
table.set_fontsize(10)

plt.subplots_adjust(left=0.1, bottom=0.3, right=0.9, top=0.9)
plt.show()
```
